### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt

# Data for the bar plot
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'blue', 'green', 'orange']  # Different colors for each bar

# Create the bar plot
plt.figure(figsize=(8, 6))  # Set the figure size
plt.bar(categories, values, color=colors)  # Create the bar plot with custom colors

# Customize the plot
plt.title("Bar Plot of Categories", fontsize=16)  # Add a title
plt.xlabel("Categories", fontsize=12)  # Label for the x-axis
plt.ylabel("Values", fontsize=12)  # Label for the y-axis

# Show the plot
plt.show()
```
