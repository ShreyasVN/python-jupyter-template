### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt

# Data for the plot
x = [0, 1, 2, 3, 4, 5]
y = [0, 1, 4, 9, 16, 25]

# Create the line plot
plt.plot(x, y, color='blue', linestyle='--', marker='o', markersize=8, label='y = x^2')

# Customizations
plt.title('Customized Line Plot', fontsize=16)
plt.xlabel('X Axis', fontsize=14)
plt.ylabel('Y Axis', fontsize=14)
plt.grid(True, linestyle=':', color='gray')
plt.legend()
plt.xlim(0, 5)
plt.ylim(0, 30)

# Display the plot
plt.show()
```
