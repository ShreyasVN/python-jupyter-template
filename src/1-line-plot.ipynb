### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 10, 100)  # x values ranging from 0 to 10 with 100 points
y = x**2  # y values as the square of x

# Create the plot
plt.figure(figsize=(8, 6))  # Set the figure size
plt.plot(x, y, label="y = x²", color="blue", linestyle="--")  # Plot the line with a label and style

# Customize the plot
plt.title("Line Plot of y = x²", fontsize=16)  # Add a title
plt.xlabel("x values", fontsize=12)  # Add label for the x-axis
plt.ylabel("y values (x²)", fontsize=12)  # Add label for the y-axis
plt.grid(True)  # Add a grid
plt.legend(fontsize=12)  # Add a legend

# Show the plot
plt.show()
```
