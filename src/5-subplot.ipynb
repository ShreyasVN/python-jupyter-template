### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the plots
x = np.arange(1, 6)
y = x**2
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]
pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(12, 10))  # 2x2 grid of subplots

# Line chart in the first subplot
axs[0, 0].plot(x, y, color='blue', label='y = x²')
axs[0, 0].set_title("Line Chart", fontsize=14)
axs[0, 0].set_xlabel("x", fontsize=12)
axs[0, 0].set_ylabel("y", fontsize=12)
axs[0, 0].legend()
axs[0, 0].grid(True)

# Bar chart in the second subplot
axs[0, 1].bar(categories, values, color=['red', 'blue', 'green', 'orange'])
axs[0, 1].set_title("Bar Chart", fontsize=14)
axs[0, 1].set_xlabel("Categories", fontsize=12)
axs[0, 1].set_ylabel("Values", fontsize=12)

# Scatter plot in the third subplot
axs[1, 0].scatter(scatter_x, scatter_y, color='purple', s=100)
for i in range(len(scatter_x)):
    axs[1, 0].text(scatter_x[i] + 0.1, scatter_y[i], f"({scatter_x[i]}, {scatter_y[i]})", fontsize=10)
axs[1, 0].set_title("Scatter Plot", fontsize=14)
axs[1, 0].set_xlabel("x", fontsize=12)
axs[1, 0].set_ylabel("y", fontsize=12)
axs[1, 0].grid(True)

# Pie chart in the fourth subplot
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', startangle=90, colors=['gold', 'skyblue', 'lightcoral', 'limegreen'])
axs[1, 1].set_title("Pie Chart", fontsize=14)

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```
