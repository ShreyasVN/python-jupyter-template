### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt

# Data for the plots
x = [0, 1, 2, 3, 4, 5]
y1 = [0, 1, 4, 9, 16, 25]   # Quadratic
y2 = [0, 1, 2, 3, 4, 5]     # Linear
y3 = [25, 16, 9, 4, 1, 0]   # Inverted Quadratic
y4 = [0, -1, -4, -9, -16, -25] # Negative Quadratic

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# First subplot
axs[0, 0].plot(x, y1, color='blue', marker='o')
axs[0, 0].set_title('Quadratic')
axs[0, 0].grid(True)

# Second subplot
axs[0, 1].plot(x, y2, color='green', linestyle='--', marker='x')
axs[0, 1].set_title('Linear')
axs[0, 1].grid(True)

# Third subplot
axs[1, 0].plot(x, y3, color='red', linestyle='-.', marker='s')
axs[1, 0].set_title('Inverted Quadratic')
axs[1, 0].grid(True)

# Fourth subplot
axs[1, 1].plot(x, y4, color='purple', linestyle=':', marker='d')
axs[1, 1].set_title('Negative Quadratic')
axs[1, 1].grid(True)

# Overall customization
fig.suptitle('Subplot Layout Example', fontsize=18)
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Display the plots
plt.show()
```
