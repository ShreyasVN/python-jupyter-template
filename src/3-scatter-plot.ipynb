### Exercise 3: Scatter Plot with Annotations

```python
import matplotlib.pyplot as plt

# Data for the scatter plot
x = [1, 2, 3, 4, 5]
y = [10, 14, 19, 23, 27]
labels = ['Point A', 'Point B', 'Point C', 'Point D', 'Point E']

# Create the scatter plot
plt.scatter(x, y, color='blue', s=100, marker='o')

# Add annotations
for i in range(len(x)):
    plt.annotate(labels[i], (x[i], y[i]), textcoords="offset points", 
                 xytext=(5, 5), ha='center', fontsize=10, color='darkred')

# Customizations
plt.title('Scatter Plot with Annotations', fontsize=16)
plt.xlabel('X Axis', fontsize=14)
plt.ylabel('Y Axis', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```
