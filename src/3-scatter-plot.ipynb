### Exercise 3: Scatter Plot with Annotations

```python
import matplotlib.pyplot as plt

# Data for the scatter plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the scatter plot
plt.figure(figsize=(8, 6))  # Set the figure size
plt.scatter(x, y, color="blue", s=100, label="Data points")  # Scatter plot with point size

# Annotate each point with its (x, y) value
for i in range(len(x)):
    plt.text(x[i] + 0.1, y[i], f"({x[i]}, {y[i]})", fontsize=10)  # Offset the annotation slightly for readability

# Customize the plot
plt.title("Scatter Plot with Annotations", fontsize=16)  # Add a title
plt.xlabel("X-axis", fontsize=12)  # Label for the x-axis
plt.ylabel("Y-axis", fontsize=12)  # Label for the y-axis
plt.grid(True)  # Add a grid for better visualization
plt.legend(fontsize=12)  # Add a legend

# Show the plot
plt.show()
```
