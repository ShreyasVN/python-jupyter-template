### Exercise 6: Histogram Customization

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data
data = np.random.normal(50, 15, 1000)  # Mean = 50, Std Dev = 15, 1000 samples

# Create the histogram
plt.hist(data, bins=20, color='skyblue', edgecolor='black', alpha=0.7)

# Customizations
plt.title('Customized Histogram', fontsize=16)
plt.xlabel('Value', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.5)

# Add vertical line for mean
plt.axvline(data.mean(), color='red', linestyle='dashed', linewidth=2, label='Mean')

# Add legend
plt.legend()

# Display the plot
plt.show()
```
