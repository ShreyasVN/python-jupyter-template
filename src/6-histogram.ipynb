### Exercise 6: Histogram Customization

```python
import matplotlib.pyplot as plt

# Data for the histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Create the histogram
plt.figure(figsize=(8, 6))  # Set the figure size
plt.hist(data, bins=5, color='skyblue', edgecolor='black', rwidth=0.8)  # Histogram with specific color and edge

# Customize the histogram
plt.title("Histogram of Data", fontsize=16)  # Add a title
plt.xlabel("Data Values", fontsize=12)  # Label for the x-axis
plt.ylabel("Frequency", fontsize=12)  # Label for the y-axis
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add a grid for the y-axis

# Show the plot
plt.show()
```
